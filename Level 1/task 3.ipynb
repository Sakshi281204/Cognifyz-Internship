```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
df = pd.read_csv('C:/Users/asus/OneDrive/Desktop/Machine learning/Sakshi Gurule (Data Science)/Dataset .csv')
head=df.head() # in this I have cheked for the first 5 rows of the dataset
print(head)
plt.figure(figsize=(10, 6))
plt.scatter(df['Longitude'], df['Latitude'], alpha=0.5)
plt.xlabel('Longitude')
plt.ylabel('Latitude')
plt.title('Geographical Distribution of Restaurants')
plt.show()
city_counts = df['City'].value_counts()
top_cities = city_counts.head(10)

plt.figure(figsize=(10, 5))
sns.barplot(x=top_cities.values, y=top_cities.index)
plt.xlabel('Number of Restaurants')
plt.ylabel('City')
plt.title('Top 10 Cities with Highest Number of Restaurants')
plt.show()
country_counts = df['Country Code'].value_counts().head(10)

plt.figure(figsize=(10, 5))
sns.barplot(x=country_counts.values, y=country_counts.index)
plt.xlabel('Number of Restaurants')
plt.ylabel('Country Code')
plt.title('Top 10 Countries by Number of Restaurants')
plt.show()
plt.figure(figsize=(10, 6))
scatter = plt.scatter(
    df['Longitude'],
    df['Latitude'],
    c=df['Aggregate rating'],
    cmap='viridis',
    alpha=0.6
)

plt.xlabel('Longitude')
plt.ylabel('Latitude')
plt.title('Restaurant Location vs Aggregate Rating')
plt.colorbar(scatter, label='Aggregate Rating')
plt.show()
df[['Latitude', 'Longitude', 'Aggregate rating']].corr()


```
