```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
df = pd.read_csv('C:/Users/asus/OneDrive/Desktop/Machine learning/Sakshi Gurule (Data Science)/Dataset .csv')
head=df.head() # in this I have cheked for the first 5 rows of the dataset
print(head)
df.columns
df.dtypes
numerical_columns = ['Longitude','Latitude','Average Cost for two','Price range','Aggregate rating','Votes'] #these are the numerical columns on which we are going to perform statistical operations
print(numerical_columns)
df[numerical_columns].describe()
categorial_columns=['Country Code','City','Cuisine'] #in categorial columns we have to identify top ciusines and cities with the highest number of restarants 
print(categorial_columns)
city=df['City'].value_counts().sort_values(ascending=False).head(10) #this line shows top 20 cities with highest number of restarants
print(city)
# Converted to data for the better visualization
city_df = city.reset_index()
city_df.columns = ['City', 'Number of Restaurants']
city_df
top_cities = city_df['Number of Restaurants'].head(10)

plt.figure(figsize=(10,5))
top_cities.plot(kind='bar')
plt.xlabel('City')
plt.ylabel('Number of Restaurants')
plt.title('Top 10 Cities with Highest Number of Restaurants')
plt.show()
# Now i have to Identify top cuisines
cuisines = df['Cuisines'].dropna().str.split(', ') #this line splits the cuisines by comma and space
all_cuisines = cuisines.explode() #this line Seperates the list of cuisines into separate rows
top_cuisines = all_cuisines.value_counts().head(10) #this line shows top 10 cuisines
print(top_cuisines)
# Visualization of top cuisines
plt.figure(figsize=(8,5))
top_cuisines.sort_values().plot(kind='barh')
plt.xlabel('Restaurant Count')
plt.ylabel('Cuisine')
plt.title('Top 10 Cuisines by Number of Restaurants')
plt.show()
```
