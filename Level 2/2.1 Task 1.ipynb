```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
df = pd.read_csv('C:/Users/asus/OneDrive/Desktop/Machine learning/Sakshi Gurule (Data Science)/Dataset .csv')
head=df.head() # in this I have cheked for the first 5 rows of the dataset
#print(head)
columns=df.columns # in this I have checked for the Required columns of the dataset like has table booking and has online delivery
print(columns)
#Now step 1 is to calculate the percentage of both the columns 
total=len(df)
Table_Booking_Percentage=(df['Has Table booking'].value_counts()/total)*100
print('Table_Booking_Percentage=',Table_Booking_Percentage)
Online_Delivery_Percentage=(df['Has Online delivery'].value_counts()/total)*100
print('Online_Delivery_Percentage=',Online_Delivery_Percentage)
#Step 2 Comapre the Averagge Ratings of Restaurants with and without Table Booking
average_ratings_table_booking=df.groupby('Has Table booking')['Aggregate rating'].mean()
print(average_ratings_table_booking) # in this i have get the values of average ratings of booking table which is with and withount table booking in the form of yes and no
#step 3 is to analyze the availability of online delivery among restaurants with different price ranges
# required columns are Has Online delivery and Price range
price_range_online_delivery=df.groupby('Price range')['Has Online delivery'].value_counts().unstack() # in this i have grouped price range and Online delivery and by using unstack converted rows in to columns for better undertsnading
print(price_range_online_delivery)  
price_range_online_delivery.plot(kind='bar',stacked=True,figsize=(8,5))
plt.title('Online Delivery Availability Across Price Ranges')
plt.xlabel('Price Range')
plt.ylabel('Number of Restaurants With and Without Online Delivery')
plt.show()
```
