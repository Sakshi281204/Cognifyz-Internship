```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
df = pd.read_csv('C:/Users/asus/OneDrive/Desktop/Machine learning/Sakshi Gurule (Data Science)/Dataset .csv')
head=df.head() # in this I have cheked for the first 5 rows of the dataset
print(head)
df.columns
# Create visualizations to represent the distribution of ratings using different charts (histogram, bar plot, etc.).
# Compare the average ratings of different cuisines or cities using appropriate visualizations.
# Visualize the relationship between various features and the target variable to gain insights.
plt.figure(figsize=(7,5))
plt.hist(df['Aggregate rating'], bins=10)
plt.xlabel('Aggregate Rating')
plt.ylabel('Number of Restaurants')
plt.title('Distribution of Aggregate Ratings')
plt.show()
# Step 2 
df_cuisine = df.copy()
df_cuisine['Cuisines'] = df_cuisine['Cuisines'].str.split(', ')
df_cuisine = df_cuisine.explode('Cuisines')

top_cuisines = (df_cuisine.groupby('Cuisines')['Aggregate rating'].mean().sort_values(ascending=False).head(10))
plt.figure(figsize=(8,5))
top_cuisines.plot(kind='bar')
plt.xlabel('Cuisine')
plt.ylabel('Average Rating')
plt.title('Top 10 Cuisines by Average Rating')
plt.show()


# step 3 Compare average ratings of different cuisines or cities
price_rating = df.groupby('Price range')['Aggregate rating'].mean()
plt.figure(figsize=(7,5))
price_rating.plot(kind='bar')
plt.xlabel('Price Range')
plt.ylabel('Average Rating')
plt.title('Average Rating by Price Range')
plt.show()
#Table booking / Online delivery vs Rating
table_booking_rating = df.groupby('Has Table booking')['Aggregate rating'].mean()
table_booking_rating.plot(kind='bar', figsize=(6,4))
plt.title('Average Rating vs Table Booking')
plt.ylabel('Average Rating')
plt.show()
```
